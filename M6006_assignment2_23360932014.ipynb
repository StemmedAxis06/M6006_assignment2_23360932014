# Correlation Analysis

To start with, we shall perform a Pearson Correlation analysis. This technique measures the strength and direction of the linear relationship between two variables. In this case, that would be our student attendance rate and our module attendance rate. Basically, it tells us how closely our data points follow a straight line.

````python
# Let's recreate all our dataframes and such to make sure everything is clean and untouched

the_attendance_of_the_modules = the_summative_df.groupby('Module Name')['Has Attended'].mean()
but_what_of_the_students = the_summative_df.groupby(['Person Code', 'Module Name'])['Has Attended'].mean().reset_index()
average_student_by_module = but_what_of_the_students.groupby('Module Name')['Has Attended'].mean()

compare_their_attendance = pd.DataFrame({
    'Module Attendance' : the_attendance_of_the_modules,
    'Average Student Attendance' : average_student_by_module
})

# Right, so now we can actually perform our Correlation Analysis
pearson_corr, pearson_p = stats.pearsonr(
    compare_their_attendance['Module Attendance'],
    compare_their_attendance['Average Student Attendance']
)

print(f"Correlation Coefficient: {pearson_corr}")
print (f"P-value: {pearson_p}")
````

Those results look good to answering our question. There is a strong positive linear correlation (Pearson Correlation Coefficient: 0.9994, P-value: 4.90e-71) between module attendance and student attendance within that module. 

This does suggest that students with low overall attendance are indeed more likely to be enrolled in modules with low attendance rates.

### HOWEVER

# Linear Regression

So, why are we even doing linear regression? Well, it provides us with more statistical analysis by not just measuring correlation, but actually quantifying how module attendance changes with student overall attendance. 

In layman's terms, it actually makes the relationship tangible in a way that we can use it for the very purpose we find statistical relationships in the first place. To model and predict. 

### Linear Regression for First Tautological Correlation Method

````python
# God Bless these imports
from scipy import stats
import numpy as np

# First, we perform our linear regression and print the results
slope, intercept, r_value, p_value, std_err = stats.linregress(
    compare_their_attendance['Module Attendance'], 
    compare_their_attendance['Average Student Attendance']
)

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")

# Now we plot it, a scatter with associated regression line
plt.figure(figsize=(10,6))
plt.scatter(
    compare_their_attendance['Module Attendance'], 
    compare_their_attendance['Average Student Attendance']
)
plt.plot(
    compare_their_attendance['Module Attendance'], 
    slope * compare_their_attendance['Module Attendance'] + intercept, 
    color='red'
)
plt.title('Module Attendance vs Average Student Attendance Within the Module')
plt.xlabel('Module Attendance')
plt.ylabel('Average Student Attendance')
plt.show()
````

And there it is, as expected. But again, of course there's a strong linear regression. After all, its possibly a question of tautology. In some ways, we're just measuring literally the same values against each other. Obviously not completely, and we shouldn't discount it entirely, but realistically, the more important measure is the following one.

### Linear Regression for More Robust Method

````python
# Superior Linear Regression ;)
slope, intercept, r_value, p_value, std_err = stats.linregress(
    john_module_student_df['Module Attendance'], 
    john_module_student_df['Average Student Overall Attendance']
)

# The results please!
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")

# Now, we simply plot it, via a scatter and a regression line
plt.figure(figsize=(10,6))
plt.scatter(john_module_student_df['Module Attendance'], john_module_student_df['Average Student Overall Attendance'])
plt.plot(john_module_student_df['Module Attendance'], slope * john_module_student_df['Module Attendance'] + intercept, color='red')

plt.title('Module Attendance vs Overall Student Academic Attendance')
plt.xlabel('Module Attendance')
plt.ylabel('Student Overall Attendance')
plt.show()
````

# Conclusion

Well, for starters, our analysis revealed an observation of an almost perfect correlation within individual modules, suggesting students' attendance closely mirrors module attendance. This was substantiated by its linear regression. 

But, once again, this is obvious. After all, if you compare the modules a student physically ATTENDS against the module itself (as an initial reading of the question might suggest), then you end up comparing apples with apples. Of course there will be a linear correlation between student attendance within the modules they attend and low-attendance in those same modules. One causes the other. 

But, when we compare a student's overall academic record against the modules they enlisted in, this is where we get a much more interesting picture (also, pretty sure that is what the question was actually asking us and I'm being annoying, but heyho).

What we found was instead a moderately positive correlation (Pearson coefficient: 0.6599, p-value: 1.86e-07) and a linear regression explaining 43.55% of the attendance variation. What might this suggest?

Well, while students with low overall academic attendance do tend to be enrolled in modules with lower attendance rates, the relationship is not so deterministic as to say they are inherently correlated. Again, there is an element of tautology to this. After all, is the module attendance rate not low due to said non-attending students not attending? Now, this isn't as causal as our previously measurement, but there is food for thought.

That said, the data is highly temporal, which does make the even moderate correlation that much more powerful, given the correlation occured through insulated, discrete years (at face value). Though again, missing factors. For example, could a module have gained a reputation from one year that carried over into other years?

To summarise, module attendance can partially predict student attendance, but as usual with analysis of this case, there are likely other deterministic factors that have weighting. This is the primary issue with comparing two classes of variables. That said, there is enough correlation that one would not entirely be wrong for saying "Students with low overall attendance tend to attend modules that have low attendance", but they would certainly be deliberately missing a much larger, unanalysed picture.
